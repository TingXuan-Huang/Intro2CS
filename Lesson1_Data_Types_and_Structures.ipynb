## Lesson 1 — Data Types and Data Structures

**Three questions drive this whole lesson:**

1. **What kind of value am I working with?** → a *data type*
2. **How should multiple values be organized?** → a *data structure*
3. **What is the difference between each data structure, and how should we use them**

> A computer cannot manipulate a real-world concept directly. We decide how to
> *represent* it using values, and how to *organize* those values so the operations
> we need are easy and reliable.

**Everyday example — a phone contact**

Your phone cannot store the actual person. Instead, it stores information about them
in an organized way so you can quickly call, text, or find a favorite contact:

```python
contact = {
    "name": "Alex",
    "phone": "555-1234",
    "favorite": True
}
```

#### Quick review — Python's built-in types and structures

````python
# INDIVIDUAL VALUES (data types)
examples = {
    "integer": 42,             # int: whole number
    "decimal": 3.14,          # float: approximate decimal
    "text": "hello",         # str: text
    "yes/no": True,           # bool: True or False
    "missing": None,          # NoneType: no known value
}

# CONTAINERS (data structures)
a_list  = [10, 20, 20]           # ordered, mutable, duplicates allowed
a_tuple = (10, 20, 20)           # ordered, immutable, duplicates allowed
a_set   = {10, 20, 20}           # unique values, membership access
a_dict  = {"name": "Ada", "age": 36}  # unique keys, key-based access

# Falsy means Python treats the value as False inside an if statement.
falsy_values = [False, None, 0, 0.0, "", [], (), {}, set()]
print("All falsy?", all(not value for value in falsy_values))
# Decimal values such as 3.14 are called floats.

# Casting means explicitly converting a value to another type.
count = int("12")              # text -> integer
price = float("19.95")         # text -> float
label = str(101)               # integer -> text
enabled = bool(1)              # integer -> boolean

# Python is dynamically typed: names can be rebound to values of new types.
value = 10
value = "ten"                  # valid Python; value now refers to a string

for name, item in examples.items():
    print(f"{name:8} -> {item!r:8} has type {type(item).__name__}")
print(type(a_list), type(a_tuple), type(a_set), type(a_dict))
````

````python
# floats are APPROXIMATIONS — computers can't store every decimal exactly
print(0.1 + 0.2)   # not exactly 0.3
print(0.1 + 0.2 == 0.3)
````

🧭 For an intro lesson, the only takeaway about floats is: *decimals are approximate*.
Never test money with `==` on raw floats; round or use integer cents in real systems.

🧭 **Falsy values — a classic beginner trap.** In an `if`, *all* of these count as False:
`0`, `0.0`, `""`, `[]`, `{}`, and `None`. So `if revenue:` will silently skip a customer
whose revenue is a legitimate `0`. When you mean "does a value exist?", test
`if revenue is not None:` — not `if revenue:`.

````python
# Demonstrate the falsy trap
revenue = 0
if revenue:
    print("counted")
else:
    print("SKIPPED even though 0 is a real value!")

# The safe check when you mean 'is a value present?'
print("present?", revenue is not None)
````

````python
# None — the absence of a known value. Different from 0, "" and False.
cancellation_date = None
print(cancellation_date)

# 0     -> a known number
# ""    -> a known (empty) string
# False -> a known boolean
# None  -> no value available yet
````

````python
# Type conversion (casting)
age_number = int("25")
price_number = float("19.99")
print(age_number + 1, price_number * 2)

# Type checking
print(type(price_number))
print(isinstance(price_number, float))
````

### Operations reveal properties

Each type *permits* some operations and *forbids* others — and the permitted operations
are the type's properties made concrete. In the cells below, watch two things: what works,
and what fails. When an operation fails, the error message names the property you assumed
wrongly.

````python
# NUMBERS support arithmetic — that is what "numeric" means
print(7 + 3, 7 - 3, 7 * 3)   # add, subtract, multiply
print(7 / 2)                 # true division ALWAYS returns a float: 3.5
print(7 // 2)                # floor division stays whole: 3
print(7 % 2)                 # remainder: 1
print(2 ** 10)               # power: 1024
print(round(3.14159, 2), abs(-5))

# The type of the RESULT is part of the operation: 10 / 2 is 5.0, not 5
print(10 / 2, type(10 / 2).__name__)
````

````python
# STRINGS are ordered sequences of characters — position operations work
word = "revenue"
print(word[0])         # first character, like a list
print(word[0:3])       # slice: 'rev'
print(word[-1])        # last character
print(len(word))       # length
print("ven" in word)   # membership test
print("Q" + "3")       # concatenation: 'Q3'
print("-" * 10)        # repetition: '----------'
````

````python
# STRINGS are IMMUTABLE — every "change" builds a NEW string
name = "alice"
try:
    name[0] = "A"
except TypeError as e:
    print("TypeError:", e)

print(name.upper())   # returns a NEW string...
print(name)           # ...the original is untouched

cleaned = "  Alice Smith \n".strip()    # new string, whitespace removed
parts = "West,East,North".split(",")    # str -> list of strings
print(cleaned, "|", parts)
````

````python
# BOOLEANS are produced by comparisons — and combined with and/or/not
revenue = 1200
print(revenue > 1000)                       # True
print(revenue == 1200, revenue != 0)
print(0 < revenue < 2000)                   # chained comparison
print(revenue > 1000 and revenue < 1500)
print(not revenue > 1000)

# bool is a subtype of int: True counts as 1 — handy for counting
checks = [True, False, True, True]
print(sum(checks), "of", len(checks), "checks passed")
````

````python
# NONE supports almost no operations — no value means nothing to operate on
missing = None
try:
    missing + 1
except TypeError as e:
    print("TypeError:", e)

# The only sensible questions about None: is it there, or not?
print(missing is None)
print(missing is not None)
````

🧭 **Why this matters:** most "data" bugs are really *type* bugs — revenue imported as
text, IDs treated as numbers, missing values stored as the string `"N/A"`, booleans stored
as `"Yes"`/`"No"`. A calculation can't be correct if the program misunderstands the values.

**Mini-exercise — pick the type:** customer name → `str`; product price → `float`;
number of purchases → `int`; subscription active → `bool`; missing cancellation date →
`None`; ZIP code → `str`; transaction ID → `str`.

**Language note:** Python is *dynamically typed* — a variable can be rebound to a different
type. In Java you must declare the type (`int value = 10;`). Even in Python the *values*
still have types; only the *variable* isn't pinned to one.

````python
# Python: a variable can be rebound to a different type
value = 10
print(type(value))
value = "ten"
print(type(value))
````

### Dynamic typing compared with C and Java

Python attaches the type to the **value**, and a variable name may later point to a value
of another type. C and Java are **statically typed**: a variable's declared type is checked
before the program runs.

**C**

~~~c
int value = 10;
value = "ten";   // compile-time error: value must remain an int
~~~

**Java**

~~~java
int value = 10;
value = "ten";   // compile-time error: String cannot be assigned to int
~~~

Static typing catches some mistakes earlier. Dynamic typing is flexible, but the values
still have types and invalid operations still fail.

## Native Python data structures
### List · Tuple · Set · Dictionary

### 3.1 List — ordered · mutable · allows duplicates · accessed by position

````python
sales = [120, 150, 150, 180]   # order matters, 150 repeats
print(sales[0])                # first item (position 0)
sales.append(200)              # add to the end
sales[1] = 160                 # change item in place
print(sales)
````

````python
# LIST operations map one-to-one onto its properties
sales = [120, 150, 150, 180]
print(sales[-1])           # ordered -> negative index counts from the end
print(sales[1:3])          # ordered -> slicing takes a sub-sequence
print(sales.count(150))    # duplicates allowed -> counting occurrences is meaningful
sales.insert(1, 130)       # mutable -> insert at any position
sales.remove(150)          # mutable -> removes the FIRST matching value
last = sales.pop()         # mutable -> take from the end
print(sales, "| popped:", last)

# sort() changes the list IN PLACE and returns None; sorted() builds a NEW list
prices = [30, 10, 20]
print(sorted(prices), prices)                # new list; original untouched
result = prices.sort()
print(prices, "| sort() returned:", result)  # mutated; returns None!
````

🧭 Use a list when order matters, items may repeat, you process items in sequence, or you
add/remove/modify elements. Business example: `monthly_revenue = [12000, 13500, 12800, 15000]`.

### 3.2 Tuple — ordered · **immutable** · allows duplicates

````python
coordinate = (47.6062, -122.3321)   # a fixed pair that belongs together
reporting_period = (2026, "Q3")
print(coordinate, reporting_period)

# Immutable: this would raise an error (uncomment to see)
# coordinate[0] = 0.0   # TypeError: 'tuple' object does not support item assignment
````

````python
# TUPLE: position operations work (ordered); change operations don't exist (immutable)
coordinate = (47.6062, -122.3321)
print(coordinate[0], coordinate[-1], coordinate[:1])   # indexing and slicing, like a list

try:
    coordinate.append(0.0)
except AttributeError as e:
    print("AttributeError:", e)   # tuples don't even HAVE append

# Immutability's payoff: unpack a fixed structure into named parts
lat, lon = coordinate
print("latitude:", lat, "| longitude:", lon)

# A tuple's entire public method list — nothing that mutates:
print([m for m in dir(coordinate) if not m.startswith("_")])
````

🧭 Use a tuple when several values belong together and should not normally change — a
coordinate, a year–month pair, a composite key. The point is *immutability + intent*, not
"a list with round brackets."

### 3.3 Set — unique values · membership-based · not position-based

````python
# Real customer IDs from the pinned UCI Online Retail classroom extract.
import csv
from pathlib import Path

with (Path("course_data") / "lesson2_transactions_base.csv").open(newline="") as file:
    retail_rows = list(csv.DictReader(file))

customer_ids = {row["customer_id"] for row in retail_rows}
one_customer = next(iter(customer_ids))
customer_ids.add(one_customer)  # duplicate ignored
print(customer_ids)

print(one_customer in customer_ids)   # fast membership test
````

````python
a = {101, 102, 103}
b = {102, 103, 104}
print("intersection:", a & b)   # in both
print("union       :", a | b)   # in either
print("difference  :", a - b)   # in a but not b
````

````python
# SET: membership operations work; position operations don't exist (no order)
ids = {103, 101, 102}
try:
    ids[0]
except TypeError as e:
    print("TypeError:", e)          # no order -> no "first element"

print(len({10, 20, 20}))            # 2 — duplicates removed at creation
print({1, 2, 3} ^ {3, 4})           # symmetric difference: in one but not both
print({1, 2} <= {1, 2, 3})          # subset test

ids.discard(999)                    # discard: silent if the value is absent
# ids.remove(999)                   # remove: KeyError if the value is absent
print(set([120, 150, 150, 180]))    # the classic dedup idiom: list -> set
````

🧭 Reach for a set when the question is: *is this present? what's unique? what's shared?*
Business example: customers reached by both email and social = `email_ids & social_ids`.

### 3.4 Dictionary — maps keys → values · accessed by key

````python
# These rows come from the pinned UCI Online Retail extract in course_data/.
# We use only a compact, anonymous classroom view of the real transactions.
import csv
from pathlib import Path

with (Path("course_data") / "lesson2_transactions_base.csv").open(newline="") as file:
    retail_rows = list(csv.DictReader(file))

revenue_by_customer = {}
for row in retail_rows:
    revenue_by_customer[row["customer_id"]] = revenue_by_customer.get(row["customer_id"], 0) + (
        int(row["quantity"]) * float(row["unit_price"])
    )

first_row = retail_rows[0]
customer = {
    "customer_id": first_row["customer_id"],
    "country": first_row["source_country"],
    "revenue": round(revenue_by_customer[first_row["customer_id"]], 2),
}
print(customer["country"])
customer["revenue"] += 100     # update by key
print(customer["revenue"])
````

````python
# A dictionary as a LOOKUP TABLE
example_customer_id = sorted(revenue_by_customer)[1]
print(revenue_by_customer[example_customer_id])
# "give me this customer ID" — not "position 2"
````

````python
# DICT: every operation is BY KEY — reading, testing, adding, deleting
customer = {"customer_id": 101, "name": "Alice", "revenue": 1200.50}

try:
    customer["email"]
except KeyError as e:
    print("KeyError:", e)                    # missing key is an error...
print(customer.get("email", "unknown"))     # ...unless you ask with a default

print("name" in customer)    # `in` tests KEYS...
print("Alice" in customer)   # ...not values, so this is False!
print(list(customer.keys()), list(customer.values()))

customer["email"] = "alice@example.com"     # plain assignment adds a NEW key
removed = customer.pop("revenue")           # pop removes a key and returns its value
print(customer, "| removed:", removed)
````

🧭 Use a dictionary when data is accessed by a *name/identifier* rather than a numeric
position. Dictionaries are the intuition behind JSON objects, database indexes, key-value
stores, caches, and API responses.

**Why can't a list be a dictionary key or a set member?** Keys and set elements must be
*immutable* (hashable). `(101, "West")` works as a key; `[101, "West"]` does not. This is
the same immutability idea from tuples — it's what makes fast key lookup possible.

````python
# Tuple key works; list key fails
lookup = {(2026, "Q3"): 15000}
print(lookup[(2026, "Q3")])
# {[1, 2]: "x"}   # TypeError: unhashable type: 'list'
````

### Structure comparison

| Structure  | Ordered | Mutable | Duplicates | Access | Good for |
|------------|---------|---------|------------|--------|----------|
| List       | Yes | Yes | Yes | Position | Ordered observations |
| Tuple      | Yes | No  | Yes | Position | Fixed grouped values |
| Set        | No (membership) | Yes | No | Membership | Unique values |
| Dictionary | Insertion-ordered | Yes | Keys unique | Key | Records & lookups |

**Choose by the operation:**
Need order → **List** · fixed grouped values → **Tuple** · uniqueness → **Set** ·
lookup by identifier → **Dictionary**.

````python
# THE SAME OPERATION ACROSS ALL FOUR — the behavior differences ARE the properties
a_list, a_tuple = [10, 20, 30], (10, 20, 30)
a_set, a_dict = {10, 20, 30}, {10: "a", 20: "b", 30: "c"}

# len() works on every container
print(len(a_list), len(a_tuple), len(a_set), len(a_dict))

# [0] means "position 0" for ordered structures, "KEY 0" for a dict, an error for a set
print(a_list[0], a_tuple[0])
print(a_dict.get(0, "no key 0"))   # dict: key lookup, not position
try:
    a_set[0]
except TypeError:
    print("set: no positions to index")

# `in` asks about values for list/tuple/set — but about KEYS for a dict
print(10 in a_list, 10 in a_tuple, 10 in a_set, 10 in a_dict, "a" in a_dict)

# + concatenates sequences; order-free structures refuse (nothing can "come after")
print(a_list + [40], a_tuple + (40,))
try:
    a_set + {40}
except TypeError:
    print("set: no + (use | for union)")
````

🧭 **When an operation fails, Python names the property you assumed wrongly.**

- `TypeError: 'str' object does not support item assignment` → you assumed mutability.
- `AttributeError: 'tuple' object has no attribute 'append'` → same assumption, different symptom.
- `TypeError: 'set' object is not subscriptable` → you assumed ordering.
- `KeyError: 'email'` → you assumed a key was present.

Read error messages as property lessons, not as failures.

## Unit 4 — Mutability, references, and copying  ·  ~30 min

> Assignment usually gives an object *another name*; it does not necessarily make an
> independent copy.

````python
# Mutable objects can change after creation
sales = [100, 200]
sales.append(300)
print(sales)

# Immutable objects cannot be changed in place — this makes a NEW string
name = "Alice"
name = name + " Smith"
print(name)
````

````python
# Aliasing: two names for the SAME list
original = [100, 200]
alias = original
alias.append(300)
print("original:", original)   # changed too!
````

````python
# Copying: make an INDEPENDENT list
original = [100, 200]
copied = original.copy()
copied.append(300)
print("original:", original)   # unchanged
print("copied  :", copied)
````

🧭 Aliasing explains many confusing bugs: a function "unexpectedly" changes the caller's
list, a "backup" that changes with the original, two variables that move together. The fix
is usually `.copy()` (or `copy.deepcopy()` for nested structures).

**Predict, then run** the next cell before looking at the output.

````python
customer_a = {"name": "Alice", "revenue": 100}
customer_b = customer_a          # alias, not a copy
customer_b["revenue"] = 200
print(customer_a["revenue"])     # what prints — 100 or 200?
````

````python
# Repair: copy so they're independent
customer_a = {"name": "Alice", "revenue": 100}
customer_b = customer_a.copy()
customer_b["revenue"] = 200
print("a:", customer_a["revenue"], "| b:", customer_b["revenue"])
````

## Unit 5 — Stacks and queues (concepts)  ·  ~30 min

````python
# STACK — Last In, First Out (LIFO). Think: undo history.
history = []
history.append("Open page")
history.append("Change filter")
history.append("Delete row")
last_action = history.pop()      # removes the most recent
print("undo:", last_action)
print("remaining:", history)
````

````python
# A STACK is not a new type — it's a LIST you promise to use from ONE end only
history = ["Open page", "Change filter", "Delete row"]
history.append("Rename column")   # push = list.append (mutable, grows at the end)
print("peek:", history[-1])       # peek = negative indexing (ordered!)
print("undo:", history.pop())     # pop  = list.pop() from the SAME end
print("empty?", not history)      # an empty list is falsy — clean emptiness test

# The list would ALLOW more (history.pop(0) is legal) — the stack is a DISCIPLINE:
# touch only one end, and "most recent first" is guaranteed by construction.
````

````python
# QUEUE — First In, First Out (FIFO). Think: a service line.
from collections import deque
requests = deque()
requests.append("Request A")
requests.append("Request B")
requests.append("Request C")
next_request = requests.popleft()   # removes the oldest
print("serve:", next_request)
print("waiting:", list(requests))
````

````python
# A QUEUE uses both ends: in at the back, out at the front
line = ["Request A", "Request B", "Request C"]
print("serve:", line.pop(0))   # this works! ...so why does deque exist?

# Property answer: a list is one contiguous block, so removing the FRONT
# shifts every remaining element left. Fine for 3 items; slow for a million.
from collections import deque
requests = deque(["Request A", "Request B", "Request C"])
print("serve:", requests.popleft())    # deque is built to be fast at BOTH ends
requests.appendleft("VIP request")     # ...which also allows jumping the line
print("Request B" in requests, len(requests), list(requests))  # still sequence-like
````

🧭 Stack uses: undo, calculator evaluation, nested function calls, matching parentheses.
Queue uses: service requests, tasks waiting for a worker, messages awaiting delivery.
Use `deque` for queues — removing the first element of a plain list forces every remaining
element to shift, which is slow.

### Hash maps and trees — two useful mental models

A **hash map** stores key → value pairs for fast lookup. Python's built-in dictionary is a
hash-map implementation, so you already used one. A hash function turns a key into a
location; Python handles collisions and resizing behind the scenes.

A **tree** stores parent–child relationships. File folders, an organization chart, and an
HTML document all have tree-like shape. Python has no single built-in Tree type because
different trees need different rules; a small node can be represented with nested dicts.

````python
# HASH MAP: retrieve a value by a unique key.
# These product prices come from the pinned UCI retail extract.
price_by_product = {
    row["product"]: float(row["unit_price"])
    for row in retail_rows[:6]
}
example_product = next(iter(price_by_product))
print(price_by_product[example_product])

# TREE: each node has a value and zero or more children
course_tree = {
    "name": "Intro to CS",
    "children": [
        {"name": "Data", "children": []},
        {"name": "Algorithms", "children": [
            {"name": "Search", "children": []}
        ]},
    ],
}

def print_tree(node, depth=0):
    print("  " * depth + node["name"])
    for child in node["children"]:
        print_tree(child, depth + 1)

print_tree(course_tree)
````

````python
# A HASH MAP is not something you build in Python — the dict IS one.

# Every dict operation from earlier IS a hash-map operation:
print(price_by_product.get("NOT A PRODUCT", "not tracked"))  # lookup with a default
print(example_product in price_by_product)                      # key membership

# Dict KEYS behave like a SET: unique, hashable — set operations work on them directly.
watchlist = set(list(price_by_product)[:2])
print(price_by_product.keys() & watchlist)  # intersection of keys and a set
# That is why keys must be immutable (tuples ok, lists not) — same rule as set members.
````

````python
# A TREE is nothing new either: dicts and lists NESTED inside each other
# Each node = a dict (named fields -> key access); children = a list (ordered -> position)
print(course_tree["children"][1]["children"][0]["name"])   # dict[key] and list[pos], chained

# Growing the tree is just list.append on the right children list
course_tree["children"][0]["children"].append({"name": "Types", "children": []})

def count_nodes(node):
    return 1 + sum(count_nodes(child) for child in node["children"])

print("nodes:", count_nodes(course_tree))
print_tree(course_tree)
# This nested dict+list shape is EXACTLY what JSON stores — see Unit 7.
````

## Unit 6 — NumPy and pandas with a real dataset  ·  ~60 min

````python
# One record = one dictionary; a dataset = a list of dictionaries.
# Build a compact customer view from the real transaction extract.
country_by_customer = {}
for row in retail_rows:
    country_by_customer[row["customer_id"]] = row["source_country"]

customers = [
    {
        "customer_id": customer_id,
        "country": country_by_customer[customer_id],
        "revenue": round(revenue, 2),
    }
    for customer_id, revenue in sorted(revenue_by_customer.items())
]
customer = customers[0]
print(customer)
print(len(customers), "records")
````

This is the same shape as a JSON array of objects, database rows, and DataFrame rows.

````python
import numpy as np

revenue = np.array([record["revenue"] for record in customers])
print("mean :", revenue.mean())
print("*1.05:", revenue * 1.05)     # applies to EVERY element at once
print("shape:", revenue.shape, "| dtype:", revenue.dtype)
````

````python
# List vs NumPy array: same-looking code, different meaning
print([1, 2, 3] * 2)                 # list repetition -> [1, 2, 3, 1, 2, 3]
print(np.array([1, 2, 3]) * 2)       # numeric multiply -> [2 4 6]
````

````python
# A NUMPY ARRAY is a list with the flexibility REMOVED — and that is the point
revenue = np.array([record["revenue"] for record in customers])

# LIKE a list: ordered and positional
print(revenue[0], revenue[1:], len(revenue))

# UNLIKE a list: one dtype only — mixing types silently coerces EVERYTHING
print(np.array([1, "two", 3.0]))    # all became strings! check dtypes after import
# UNLIKE a list: fixed size — no .append method; np.append builds a NEW array
grown = np.append(revenue, revenue.mean())
print(len(revenue), len(grown))      # original unchanged — tuple-like thinking

# The PAYOFF of the rigidity: whole-array operations no native container has
print(revenue * 1.05)             # elementwise math
print(revenue > revenue.mean())   # elementwise comparison -> boolean array
print(revenue[revenue > revenue.mean()])  # boolean mask: filter in one expression
````

🧭 A Python list is a flexible general container. A NumPy array is a *numerical* structure
built to apply math efficiently across many values (homogeneous, has a shape and dtype).

🧭 **Abstract structures are native structures plus a rule.**

| Pattern | Built from | Property added / removed | Typical use |
|---|---|---|---|
| Stack | a list, used at one end only | adds LIFO discipline | undo, call tracking, matching brackets |
| Queue | a deque (list-like, fast at both ends) | adds FIFO fairness | service lines, task processing |
| Hash map | the dict — it *is* one | key-based lookup; keys act like a set | find a record by ID |
| Tree | dicts nested inside lists | adds hierarchy (one parent per node) | file systems, org charts, JSON/HTML |
| NumPy array | a homogeneous, fixed-size "list" | removes mixed types; adds elementwise math | bulk numeric work |

New words, same four building blocks. Choosing a structure means choosing which
operations you want cheap — and which you want forbidden.

````python
import pandas as pd

df = pd.DataFrame(customers)
print(df)
print("mean revenue:", df["revenue"].mean())
print(df.dtypes)
````

🧭 A DataFrame is a 2-D table: labeled rows and columns, different types per column, like a
spreadsheet or SQL table. It adds column names, indexing, alignment, missing-value handling
(`None` becomes `NaN`), and vectorized operations that a plain list of dicts does not have.

**When to use what:** list → ordered general collection · dictionary → record or lookup ·
NumPy array → numerical computation · DataFrame → labeled tabular analysis ·
database table → persistent, shared, queryable data.

**Representation chain (preview only):**
`Python scalar → List → NumPy array → Tensor` (tensors = multidimensional numeric arrays
for ML; we don't operate on them yet).

````python
# Missing values flow into DataFrames as NaN
df2 = pd.DataFrame([
    {"customer_id": 101, "cancellation_date": None},
    {"customer_id": 102, "cancellation_date": "2026-05-01"},
])
print(df2)
print(df2["cancellation_date"].isna())
````

### A real dataset: Fisher's Iris flower measurements

The Iris dataset contains real measurements of 150 flowers. It ships with scikit-learn,
so this lesson remains runnable without downloading a file. Watch how each operation
connects to a structural property rather than memorizing pandas methods.

````python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
flowers = iris.frame.rename(columns={"target": "species_code"})
flowers["species"] = flowers["species_code"].map(dict(enumerate(iris.target_names)))

print(flowers.head())
print("shape:", flowers.shape)       # 2D: rows × columns
print("column types:\n", flowers.dtypes)
print("duplicate rows:", flowers.duplicated().sum())
print("missing values:\n", flowers.isna().sum())
````

**Property check:** shape tells us this is a 2D table. Columns are labeled, so we access
them by name. Each numeric column is homogeneous, enabling vectorized math. Species is a
nominal category: grouping and counting make sense, but averaging species names does not.

````python
# POSITIONAL access: first three rows, first two columns
print(flowers.iloc[:3, :2])

# LABEL access + boolean filtering
wide_setosa = flowers.loc[
    (flowers["species"] == "setosa") & (flowers["sepal width (cm)"] > 3.5),
    ["sepal length (cm)", "sepal width (cm)", "species"],
]
print(wide_setosa.head())

# Vectorized operation: NumPy performs one calculation across a homogeneous array
petal_area = flowers["petal length (cm)"].to_numpy() * flowers["petal width (cm)"].to_numpy()
print("estimated petal areas:", petal_area[:5])

# Grouping uses the nominal species label; mean is applied only to measurements
summary = flowers.groupby("species")[["sepal length (cm)", "petal length (cm)"]].mean()
print(summary)
````

### The same data, different structures

- A row converted to a dictionary becomes a named record.
- A column converted to a list becomes an ordered general-purpose sequence.
- Species converted to a set answers “which unique species exist?”
- Measurements converted to a NumPy array become a homogeneous numeric matrix.

````python
one_flower = flowers.iloc[0].to_dict()
length_sequence = flowers["sepal length (cm)"].to_list()
unique_species = set(flowers["species"])
measurement_matrix = flowers.iloc[:, :4].to_numpy()

print("dict record:", one_flower)
print("first list values:", length_sequence[:5])
print("set membership:", "setosa" in unique_species)
print("NumPy shape/dtype:", measurement_matrix.shape, measurement_matrix.dtype)
````

🧭 JSON mirrors nested dictionaries and lists — great for APIs, nested records, config, and
data exchanged between systems.

````python
import io, pandas as pd
csv_text = "customer_id,name,revenue\n101,Alice,1200.50\n102,Bob,850.00\n"
print(csv_text)
df = pd.read_csv(io.StringIO(csv_text))
print(df)
````

🧭 CSV is simple, rectangular, and universally supported — but it does **not** preserve type
information reliably. A ZIP code like `"00123"` can be read back as the number `123`, losing
the leading zeros. JSON and databases preserve types better.

### Match the storage format to the structure

| In-memory idea | Natural storage | What is preserved | What may be lost |
|---|---|---|---|
| 2D DataFrame | CSV | rows, columns, text values | exact types, indexes, constraints |
| nested dict/list tree | JSON | nesting, keys, scalar categories | tuples, sets, custom types |
| homogeneous nD array | NPY | shape and exact NumPy dtype | easy use outside NumPy |
| typed analytical table | Parquet | column types, nulls, compact columns | human readability |
| shared keyed records | database | schema, unique keys, indexes | simple file portability |

Storage is not neutral: serialization converts an in-memory structure into bytes, and the
chosen format decides which properties can be reconstructed later.

````python
# CSV type loss: leading zeros disappear unless we specify a dtype.
zip_csv = "customer_id,zip_code\nC01,00123\nC02,02139\n"
guessed = pd.read_csv(io.StringIO(zip_csv))
protected = pd.read_csv(io.StringIO(zip_csv), dtype={"customer_id": "string", "zip_code": "string"})
print("guessed types:\n", guessed)
print("protected types:\n", protected)

# JSON supports lists and dictionaries, but sets require an explicit conversion.
record = {"customer_id": "C01", "tags": list({"new", "email"})}
print(json.dumps(record))

# NumPy's binary representation keeps dimensionality and dtype.
import io as binary_io
buffer = binary_io.BytesIO()
np.save(buffer, measurement_matrix)
buffer.seek(0)
restored = np.load(buffer)
print("restored shape/dtype:", restored.shape, restored.dtype)
````

### Final decision checklist

For any value or container, ask:

1. What does each value mean, and which operations are legitimate?
2. What is the shape: scalar, sequence, table, or higher-dimensional array?
3. Does order matter? Can it change? Must values be unique?
4. Will we access by position, label/key, membership, or parent–child path?
5. Must values share a type for efficient numerical work?
6. Which file format preserves the properties we need tomorrow?
